### REASON FOR USING THE REPLACE NULL VALUES USING MEAN
--------
##### The mean is the most frequent used measure of central tendency. Since the data set give is missing at most three values from each column and the mean gives the average of the data set values it is best to replace null values with the mean because it give the best accuracy.

```python
mean_1=book_final['fixed acidity'].mean()

adove_mean_count= book_final[book_final['fixed acidity']>mean_1]['fixed acidity'].count()
below_mean_count= book_final[book_final['fixed acidity']<mean_1]['fixed acidity'].count()

labels=['Greater than Mean','Less than Mean']
sizes=[adove_mean_count,below_mean_count]
colors=['yellow','green']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%',startangle=60)
plt.axis('equal')
plt.title('Fixed Acidity Greather than and Less than Mean')
plt.show()
```
